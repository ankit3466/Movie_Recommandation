in this we are using angular distance between two movies to find how similar they are.

```python
import sklearn
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
## countvectorizer function is used to count the frequencies of words in a text
cv = CountVectorizer()
```

```python
text = ["london paris london","paris paris london"]
text
```

```python
x = cv.fit_transform(text)
```

```python
x.toarray()
```

```python
## 1 row of array [2,1] means that london occurs 2 times and paris occur 1 time and same for second row for 2 text 
```

```python
## now to find the similarity between this two text
```

```python
similer = cosine_similarity(x)
```
